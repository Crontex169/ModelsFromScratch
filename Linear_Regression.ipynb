```python
import numpy as np
```

```python
class LineerReg:
  def __init__(self,lr=0.0005):
    self.lr=lr
    self.w=None
    self.b=None
  def forward(self,X):
    X= (X@self.w) +self.b
    return X
  def initilaze(self,X):
    n_features=X.shape[1]
    self.w=np.random.randn(n_features,1)*0.01
    self.b=0
  def backward(self,X,y,output):
    self.cost = np.sum((output - y)**2) / (2 * X.shape[0])
    self.dw=(X.T @ (output-y))/X.shape[0]
    self.db=np.sum((output-y))/X.shape[0]
    self.w=self.w-self.lr*self.dw
    self.b=self.b-self.lr*self.db
  def fit(self,X,y,iterations=100):
    assert isinstance(X,np.ndarray) and isinstance(y,np.ndarray)
    assert isinstance(iterations,int)
    assert X.shape[0]==y.shape[0]
    self.initilaze(X)
    for _ in range(iterations):
      output=self.forward(X)
      self.backward(X,y,output)
      if _ % 100 == 0:
          print(f"Iteration {_}, Cost: {self.cost}")
  def predict(self,X):
    return self.forward(X)

Linear=LineerReg(lr=0.0005)
X=np.array([[24.0,50.0,15.0,38.0,87.0]])
y=np.array([[21.549452,47.464463,17.218656,36.586398,87.288984]])
X=X.T
y=y.T
Linear.fit(X,y,100000)
print(Linear.predict(X))


```
